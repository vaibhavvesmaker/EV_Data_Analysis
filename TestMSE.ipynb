```python
import pandas as pd
import random

# Generate dummy data for fleet performance analytics
data = {
    'Vehicle_ID': [f'V{str(i).zfill(3)}' for i in range(1, 101)],
    'Fleet_Type': [random.choice(['Truck', 'Van', 'SUV', 'Sedan']) for _ in range(100)],
    'Distance_Travelled_km': [random.randint(500, 5000) for _ in range(100)],
    'Fuel_Consumed_L': [random.uniform(50, 500) for _ in range(100)],
    'Maintenance_Cost_USD': [random.uniform(100, 2000) for _ in range(100)],
    'Last_Service_Date': [pd.Timestamp('2025-01-01') + pd.to_timedelta(random.randint(0, 90), unit='d') for _ in range(100)],
    'Driver_ID': [f'D{str(i).zfill(3)}' for i in range(1, 101)]
}

dummy_data = pd.DataFrame(data)
dummy_data.to_csv('fleet_performance_dummy_data.csv', index=False)

```
